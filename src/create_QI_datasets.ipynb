```python
import const
import pandas as pd
```

```python
SQUAD_DS_TRAINING_PATH = const.DATASETS_FOLDER + "squad-train-v2.0.csv"
ULTRACHAT_DS_PATH = const.DATASETS_FOLDER + "ultrachat_50k.csv"
REDDIT_DS_PATH = const.DATASETS_FOLDER + "reddit_ds.csv"
```

```python
SQUAD_DS_TESTING_PATH = const.DATASETS_FOLDER + "squad-test-v2.0.csv"
```

```python
squad_train_ds = pd.read_csv(SQUAD_DS_TRAINING_PATH)
squad_test_ds = pd.read_csv(SQUAD_DS_TESTING_PATH)
ultrachat_ds = pd.read_csv(ULTRACHAT_DS_PATH)
reddit_ds = pd.read_csv(REDDIT_DS_PATH)
```

```python
squad_train_ds.columns
```

```python
squad_train_ds['is_searchable'] = 1.0
squad_test_ds['is_searchable'] = 1.0
```

```python
ultrachat_ds.columns
```

```python
split_ratio = 0.7
split_ind = int(len(ultrachat_ds) * split_ratio)
ultrachat_train_ds = ultrachat_ds[0:split_ind]
ultrachat_test_ds = ultrachat_ds[split_ind:]
```

```python
training_qi_ds = pd.concat([squad_train_ds, ultrachat_train_ds], ignore_index=True)
```

```python
reddit_ds['question'] = reddit_ds['0']
reddit_ds['is_searchable'] = 0.0
reddit_ds = reddit_ds[['question', 'is_searchable']]
```

```python
split_ratio = 0.7
split_ind = int(len(reddit_ds) * split_ratio)
reddit_train_ds = reddit_ds[0:split_ind]
reddit_test_ds = reddit_ds[split_ind:]
```

```python
training_qi_ds = pd.concat([training_qi_ds, reddit_train_ds], ignore_index=True)
```

```python
testing_qi_ds = (pd.concat([squad_test_ds, ultrachat_test_ds, reddit_test_ds],
   ignore_index=True))
```

```python
training_qi_ds.to_csv(const.DATASETS_FOLDER + "QI_training.csv")
testing_qi_ds.to_csv(const.DATASETS_FOLDER + "QI_testing.csv")
```
